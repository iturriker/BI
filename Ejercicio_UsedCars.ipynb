```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el DataFrame
data = pd.read_csv('/content/UsedCars.csv')
```

```python
# Configurar el estilo de las gr치ficas
sns.set_palette("husl")
```

# 1. Gr치fico de barras: Precio promedio por marca



```python
# prompt: gr치fico de barras: precio promedio por marca

# Calcular el precio promedio por marca
average_price_by_brand = data.groupby('Brand')['Price'].mean()

# Crear el gr치fico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=average_price_by_brand.index, y=average_price_by_brand.values)
plt.xlabel('Marca')
plt.ylabel('Precio Promedio')
plt.title('Precio Promedio por Marca')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x
plt.tight_layout()
plt.show()
```

# 2. Gr치fico de dispersi칩n: Relaci칩n entre Kilometraje y Precio



```python
# prompt: Gr치fico de dispersi칩n: Relaci칩n entre Kilometraje y Precio

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Mileage', y='Price', data=data, hue='Brand')
plt.title('Relaci칩n entre Kilometraje y Precio')
plt.xlabel('Kilometraje')
plt.ylabel('Precio')
plt.show()
```

# 3. Gr치fico de caja: Distribuci칩n de Precios por Tipo de Carrocer칤a



```python
# prompt: Gr치fico de caja: Distribuci칩n de Precios por Tipo de Carrocer칤a

plt.figure(figsize=(12, 6))
sns.boxplot(x='Body', y='Price', data=data)
plt.xlabel('Body')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 4. Gr치fico de viol칤n: Distribuci칩n del Tama침o del Motor por Marca

```python
# prompt: Gr치fico de viol칤n: Distribuci칩n del Tama침o del Motor por Marca

plt.figure(figsize=(12, 6))
sns.violinplot(x='Brand', y='EngineV', data=data)
plt.xlabel('Brand')
plt.ylabel('EngineV')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# 5. Gr치fico de calor: Correlaci칩n entre variables num칠ricas

```python
# prompt: Gr치fico de calor: Correlaci칩n entre variables num칠ricas

# Get numeric columns
data_numeric_cols = data.select_dtypes(include=['number']).columns

# Calculate the correlation matrix
correlation_matrix = data[data_numeric_cols].corr()

# Create the heatmap
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
