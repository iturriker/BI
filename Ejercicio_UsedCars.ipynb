```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar el DataFrame
data = pd.read_csv('/content/UsedCars.csv')
```

```python
# Configurar el estilo de las gráficas
sns.set_palette("husl")
```

```python
# prompt: gráfico de barras: precio promedio por marca

# Calcular el precio promedio por marca
average_price_by_brand = data.groupby('Brand')['Price'].mean()

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x=average_price_by_brand.index, y=average_price_by_brand.values)
plt.xlabel('Marca')
plt.ylabel('Precio Promedio')
plt.title('Precio Promedio por Marca')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x
plt.tight_layout()
plt.show()
```

```python
# prompt: Gráfico de dispersión: Relación entre Kilometraje y Precio

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Mileage', y='Price', data=data, hue='Brand')
plt.title('Relación entre Kilometraje y Precio')
plt.xlabel('Kilometraje')
plt.ylabel('Precio')
plt.show()
```

```python
# prompt: Gráfico de caja: Distribución de Precios por Tipo de Carrocería

plt.figure(figsize=(12, 6))
sns.boxplot(x='Body', y='Price', data=data)
plt.xlabel('Body')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: Gráfico de violín: Distribución del Tamaño del Motor por Marca

plt.figure(figsize=(12, 6))
sns.violinplot(x='Brand', y='EngineV', data=data)
plt.xlabel('Brand')
plt.ylabel('EngineV')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# prompt: Gráfico de calor: Correlación entre variables numéricas

# Get numeric columns
data_numeric_cols = data.select_dtypes(include=['number']).columns

# Calculate the correlation matrix
correlation_matrix = data[data_numeric_cols].corr()

# Create the heatmap
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
